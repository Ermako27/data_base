```python
import numpy as np
import pandas as pd
from math import pow
```

```python
df = pd.read_csv('data/athlete_events.csv', sep=',')
```

```python
df[:20]
```

```python
# PARTICIPANTS TABLE PREPARE
participants = df.loc[:, 'Name':'Team'] # взять часть столбцов из таблицы (срез по столбцам)
participants[:20]
```

```python
res = participants.drop_duplicates(subset='Name', keep='last') # удалить дубликаты по столбцу 
res
```

```python
# GAMES TABLE PREPARE
games = df.loc[:, 'Year':'City']
res_games = games.drop_duplicates()
res_games
```

```python
# EVENTS TABLE PREPARE
events = df.loc[:, 'Sport':'Event']
events = events.drop_duplicates(subset='Event', keep='last')
events
```

```python
list_of_events = events['Event'].values
count_of_participants_in_event = [df[df['Event'] == event]['Name'].count() for event in list_of_events]
events.insert(loc=len(events.columns), column='PCount', value=count_of_participants_in_event)
events
```
